```python
from warnings import filterwarnings
filterwarnings("ignore")

import tensorflow as tf
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
# from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
```

|x1|x2|y|
|---|---|---|
|0|0|0|
|1|0|1|
|0|1|1|
|1|1|1|

```python
data = np.array([
    [0,0],
    [1,0],
    [0,1],
    [1,1]
])
```

```python
label = np.array([0,0,0,1])
```

```python
# 모델구성
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1,input_shape=(2,)))
# model = Sequential()
# model.add(Dense(1,input_shape=(2,)))
```

```python
# 컴파일
model.compile(
    optimizer = "sgd",
    # optimizer = SGD()
    loss='mse',
    # loss = MSE()
    metrics = ["acc"]
)
```

```python
model.summary()
```

```python
# 모델학습
model.fit(data,label,epochs=500)
```

```python
model.get_weights()
```

```python
# 모델평가
model.evaluate(data, label)
```

```python
# 모델예측
result = model.predict(data)
print(result)
```

```python
data = np.array([
    [0,0],
    [1,0],
    [0,1],
    [1,1]
])

label = np.array([0,0,0,1])
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1,input_shape=(2,)))
```

```python
model.compile(
    optimizer = "sgd",
    # optimizer = SGD()
    loss='mse',
    # loss = MSE()
    metrics = ["acc"]
)
```

```python
model.fit(data,label,epochs=500,verbose=0)
```

```python
model.get_weights()
```

```python
model.evaluate(data,label)
```

```python
print(model.predict(data))
```

### 3. xor 게이트

```python
data = np.array([
    [0,0],
    [1,0],
    [0,1],
    [1,1]
])

label = np.array([0,1,1,0])
```

```python
model = Sequential()
model.add(Dense(32,input_shape=(2,),activation="relu"))
model.add(Dense(1, activation='sigmoid'))

model.compile(
    optimizer=SGD(),
    # loss=MSE,
    loss='mse',
    metrics = ['acc']
)
```

```python
model.summary()
```

```python
model.fit(data,label,epochs=500,verbose=1)
```

```python
model.evaluate(data,label)
```
