```python
import cv2
print(cv2.__version__)
```

```python
img = cv2.imread('img/sheet.png')
print(img.shape)
cv2.imshow("image", img)
cv2.waitKey(0) # esc key
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    # ret -> true, false 반환
    if not ret:
        break

    cv2.imshow("camera", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, img = cap.read()
    if not ret:
        break

    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    _, img_gray1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
    _, img_gray2 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)

    cv2.imshow("cam0", img)
    cv2.imshow("camera", img_gray)
    cv2.imshow("cam1", img_gray1)
    cv2.imshow("cam2", img_gray2)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, img = cap.read()
    if not ret:
        break

    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    _, img_gray1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
    img_adaptive = cv2.adaptiveThreshold(
        img_gray, 255,
        cv2.ADAPTIVE_THRESH_MEAN_C,
        cv2.THRESH_BINARY, 15, 0)

    cv2.imshow("cam0", img)
    cv2.imshow("camera", img_gray)
    cv2.imshow("cam1", img_gray1)
    cv2.imshow("cam2", img_adaptive)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# img = cv2.imread("img/numbers.jpg")
img = cv2.imread("img/sheet.png")

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

blur = cv2.GaussianBlur(gray, (9,9),0)

_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

img_adaptive = cv2.adaptiveThreshold(
        gray, 255,
        cv2.ADAPTIVE_THRESH_MEAN_C,
        cv2.THRESH_BINARY, 15, 0)

img_adaptive1 = cv2.adaptiveThreshold(
        gray, 255,
        cv2.ADAPTIVE_THRESH_MEAN_C,
        cv2.THRESH_BINARY, 15, 0)

cv2.imshow("pic1", img)
# cv2.imshow("pic2", blur)
cv2.imshow("pic2", gray)
cv2.imshow("pic3", binary)
cv2.imshow("pic4", img_adaptive)
cv2.imshow("pic5", img_adaptive1)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
