```python
from wordcloud import WordCloud
# pip install wordcloud
import matplotlib.pyplot as plt

text = """
커리어 커리어 커리어
자존감문제 자존감문제
취업 취업 취업 취업
스트레스 스트레스
불안
진로
자기계발
"""

font_path = "C:/Windows/Fonts/malgun.ttf"  # 윈도우 기준(한글 적용)

wc = WordCloud(
    font_path=font_path,
    background_color="white",
    width=800,
    height=600,
    prefer_horizontal=0.6,   # 세로 단어 허용
    colormap="Blues"
).generate(text)

plt.figure(figsize=(8, 6))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
