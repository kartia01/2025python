```python
from wordcloud import WordCloud
# pip install wordcloud
import matplotlib.pyplot as plt

text = """
  
 
   
 



"""

font_path = "C:/Windows/Fonts/malgun.ttf"  #  ( )

wc = WordCloud(
    font_path=font_path,
    background_color="#EAF1FF",
    width=800,
    height=600,
    prefer_horizontal=0.6,   #   
    colormap="gray"
).generate(text)

plt.figure(figsize=(8, 6))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
