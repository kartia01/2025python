```python
X_train = X[:35]
y_train = y[:35]
X_test = X[35:]
y_test = y[35:]
```

```python
plt.scatter(X_train[:,0], X_train[:,1])
plt.scatter(X_test[:,0], X_test[:,1])
plt.show
```

```python
kn = KNeighborsClassifier()
kn.fit(X_train,y_train)
```

```python
new1 = [[9.8, 6.7]]
prediction = kn.predict(new1)

if prediction[0] == 1:
    print("도미입니다.")
else:
    print("방어입니다.")

print("테스트 정확도 : ", kn.score(X_test, y_test))
```
