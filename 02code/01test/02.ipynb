```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

tf.random.set_seed(0)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/surgery1.csv", header=None)
df
```

```python
X = df.iloc[:, 0:16]
y = df.iloc[:, 16]
```

```python
df.rename(columns={16: 'label'}, inplace=True)
df
```

```python
df['label'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['label'].value_counts().plot(kind='bar')
```

```python
model = Sequential()
model.add(Dense(30, input_dim=16,activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['acc']
    )
```

```python
history = model.fit(X, y, epochs = 50, batch_size=16)
```

```python
model.evaluate(X, y)
```

```python
his_dict = history.history
# his_dict
loss = his_dict['loss']
acc = his_dict['acc']

epochs = range(1,len(loss)+1)
fig = plt.figure(figsize=(10,5))

# loss 그래프
ax1 = fig.add_subplot(1,2,1)
ax1.plot(epochs,loss,color="red",label='loss')
ax1.set_title('loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()

# acc 그래프
ax2 = fig.add_subplot(1,2,2)
ax2.plot(epochs,acc,color="blue",label='acc')
ax2.set_title('acc')
ax2.set_xlabel('epochs')
ax2.set_ylabel('acc')
ax2.legend()


plt.show()
```
