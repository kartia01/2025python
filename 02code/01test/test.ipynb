````python
# 주사위 굴리기
import random

dice1 = random.randint(1,6)
dice2 = random.randint(1,6)
print(f'주사위1: {dice1} 주사위2: {dice2}')

if dice1 > dice2:
    print('주사위1 승')
elif dice1<dice2:
    print('주사위2 승')
else:
    print('무승부')
````
