### 동물별 평균체온과 최대 체중구하기

```python
import numpy as np

# 체온, 심박수, 체중(kg)
data = np.array([
    [38.5, 80, 35], # 개
    [39.1, 90, 20], # 고양이
    [36.8, 60, 400] # 말
])

ani = np.array(["개", "고양이", "말"])

# 체온, 심박수, 체중 평균값
avg_value = data.mean(axis=0)

print(avg_value)

weight = data[:,2]
print(weight)

max_weight = data[:,2].max()
print(max_weight)

max_ani = ani[data[:,2].argmax()]
print(max_ani)

print("동물별 평균수치 : ", avg_value )
print("가장 무거운 동물 : ", max_ani, f'{max_weight}kg' )
```

### 곤충 [날개수, 다리수, 몸길이(cm)]

```python
data = np.array([
    [2, 6, 1.2],
    [4, 6, 3.4],
    [0, 6, 0.8],
    [0, 8, 2.0]
])

insects = np.array(["나비","잠자리", "개미", "거미"])

# 다리가 6개인 곤충만 추출
legs = data[data[:,1] == 6]
print(legs)

# 평균 몸길이 구하기
mean_length = data[:,2].mean()
print(mean_length)
print("곤충 평균 몸길이 : ",round(mean_length,2),"cm")

# 날개가 없는 곤충출력
no_wings = insects[data[:,0] == 0]
print(no_wings)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 100)
# -1 부터 1까지 100개의 점으로 나눈다.(곡선을 만들기 위해)
# linspace(x,y,z) x=시작, y=끝, z=생성할 점의 개수
# print(x)
y = x**2

plt.figure(figsize=(5,3))
plt.plot(x,y, label="y=x^2")
plt.legend()
plt.show()
```

```python
nation = ["kor","usa", "japan", "france"]
men = [175.5,176.9,172.1,178.6]
women = [163.2,163.3,158.5,164.5]

plt.figure(figsize=(5,3))
plt.bar(nation,men)
plt.bar(nation,women)
plt.show()
```

```python
nation = ["kor","usa", "japan", "france"]
men = np.array([175.5,176.9,172.1,178.6])
women = [163.2,163.3,158.5,164.5]

ind = np.arange(len(nation))
w = 0.1

plt.figure(figsize=(5,3))
plt.bar(ind - 0.05,men,width=w, label="men", color="orange")
plt.bar(ind + 0.05,women,width=w, label="women", color="blue")
plt.xticks(ind,nation)
plt.legend()
plt.show()
```

### 출생아 수 시각화(예제)

```python
year = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
m = [248958, 223883,223356,224906,208064,184308,167686,155416,139362,133516]
g = [235592,212572,212079,213514,198179,173463,159136,147260,132975,127046]
```

```python
ind = np.arange(len(year))

plt.bar(ind-0.2, m, color='b', width=0.4)
```

```python
ind = np.arange(len(year))

plt.bar(ind+0.2, g, color='r', width=0.4)
```

```python


ind = np.arange(len(year))

w = 0.4
plt.figure(figsize=(6,3.5))
plt.bar(ind - w/2, m, width=w, color='b', label="m")
plt.bar(ind + w/2, g, width=w, color='r', label="g")
plt.xticks(ind, year)
plt.legend()
plt.xlabel('year')
plt.ylabel('number of births')
plt.show()
```

```python
year = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
m = [248958, 223883,223356,224906,208064,184308,167686,155416,139362,133516]
g = [235592,212572,212079,213514,198179,173463,159136,147260,132975,127046]

ind = np.arange(len(year))

plt.figure(figsize=(6, 3.5))
plt.bar(ind, m, color='b', label='m')
plt.bar(ind, g, bottom=m, color='r', label='g')
plt.xticks(ind, year)
plt.legend()
plt.xlabel('year')
plt.ylabel('number of births')
plt.show()
```
