```python
import matplotlib.pyplot as plt

year = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
seoul = [12.1,12,12.2,12.5,13.4,13.6,13.6,13,12.9,13.5,13.2,13.7,13.2]
jeju = [15.6,15.6,15.7,16.5,16.2,16.7,17,16.8,16.6,16.8,16.7,17.5,17]

plt.figure(figsize=(5,4))
plt.plot(year,seoul,label='Seoul')
plt.plot(year,jeju,label='Jeju')
plt.title('years temp mean')
plt.legend()
plt.xlabel('years')
plt.ylabel('temp')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

year = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
seoul = [12.1,12,12.2,12.5,13.4,13.6,13.6,13,12.9,13.5,13.2,13.7,13.2]
jeju = [15.6,15.6,15.7,16.5,16.2,16.7,17,16.8,16.6,16.8,16.7,17.5,17]

year = np.array(year)
width = 0.2

plt.figure(figsize=(5,4))
plt.bar(year - width/2,jeju,width=width,label='Jeju')
plt.bar(year + width/2,seoul,width=width,label='Seoul')
plt.title('years temp mean')
plt.legend()
plt.xlabel('years')
plt.ylabel('temp')
plt.show()
```

```python
import csv

tcount = 0
count = 0
htemp = []

# cp949 or utf-8
with open('data/seouldata2.csv',encoding='cp949') as tfile:
    data = csv.reader(tfile)
    next(data)
    for row in data:
        tcount += 1
        if row[-1] != '':
            count += 1
            htemp.append(float(row[-1]))

print(f'데이터 : 전체데이터 총 {tcount}개 / 최고 온도 데이터 수 {count} / {tcount - count}')

plt.plot(htemp)
plt.show()
```

```python
plt.hist(htemp,bins=100)
plt.show()
```
